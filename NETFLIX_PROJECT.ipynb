## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

“There is a significant association between content type (Movie/TV Show) and rating category.”

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H₀):**
There is no significant association between content type (Movie/TV Show) and rating category.

**Alternative Hypothesis (H₁):**
There is a significant association between content type (Movie/TV Show) and rating category.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import chi2_contingency

# Create contingency table
contingency_table = pd.crosstab(df['type'], df['rating'])

# Perform Chi-Square Test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

p_value
````

##### Which statistical test have you done to obtain P-Value?

Chi-Square Test of Independence

##### Why did you choose the specific statistical test?

The Chi-Square Test of Independence was chosen because both variables (content type and rating) are categorical in nature.
This test determines whether there is a statistically significant association between two categorical variables.
Since we are analyzing the relationship between content type and rating category, Chi-Square is the appropriate statistical method.

#### Conclusion for Hypothetical Statement – 1

The obtained p-value is significantly less than the significance level of 0.05.
Therefore, we reject the null hypothesis.

This result confirms that there is a statistically significant association between content type (Movie/TV Show) and rating category on Netflix.
Hence, content ratings differ based on the type of content.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H₀):**  
There is no significant difference in the release year of content added recently and content added earlier on Netflix.

**Alternative Hypothesis (H₁):**  
Content added recently on Netflix has a significantly higher release year than content added earlier.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind

# Split data into two groups
recent_content = df[df['year_added'] >= 2019]['release_year'].dropna()
older_content = df[df['year_added'] < 2019]['release_year'].dropna()

# Perform Independent t-test
t_stat, p_value = ttest_ind(recent_content, older_content, equal_var=False)

p_value
````

##### Which statistical test have you done to obtain P-Value?

Independent Samples t-test

##### Why did you choose the specific statistical test?

The Independent Samples t-test was chosen because the release year is a numerical variable and we are comparing the means of two independent groups: recently added content and older added content.
This test helps determine whether the difference in average release year between the two groups is statistically significant.

#### Conclusion for Hypothetical Statement – 2

The obtained p-value is significantly less than the significance level of 0.05.
Therefore, we reject the null hypothesis.

This result confirms that content added recently on Netflix has a significantly higher release year compared to content added earlier.
This indicates that Netflix strategically prioritizes adding newer content to its platform.

## ***6. Feature Engineering & Data Pre-processing***

### 1. Handling Missing Values

````python
# Handling Missing Values & Missing Value Imputation
# Handling missing values in categorical columns
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Not Rated', inplace=True)

# Handling missing values in numerical columns
df['year_added'].fillna(df['year_added'].mode()[0], inplace=True)
df['month_added'].fillna(df['month_added'].mode()[0], inplace=True)

# Verify missing values after imputation
df.isnull().sum()
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

Categorical missing values were handled using constant value imputation by replacing them with labels such as 'Unknown' and 'Not Rated'.
This approach preserves all records while clearly indicating missing information without introducing bias.

Numerical missing values were handled using mode imputation, which is suitable when data is skewed and helps retain the most frequent value.
These techniques were chosen to maintain dataset size, ensure consistency, and avoid loss of important information.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
# Outlier detection using IQR method for numerical columns
Q1 = df['release_year'].quantile(0.25)
Q3 = df['release_year'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap outliers
df['release_year'] = df['release_year'].clip(lower_bound, upper_bound)
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Outliers were handled using the Interquartile Range (IQR) method.
Values falling outside 1.5 times the IQR were capped within acceptable limits instead of removing records.

This approach was chosen to reduce the influence of extreme values while preserving all data points.
Capping helps maintain dataset integrity and prevents distortion in statistical analysis and machine learning models.

````python
# Expand Contraction
contractions = {"can't": "cannot", "won't": "will not", "n't": " not", "'re": " are", "'s": " is"}

def expand_contractions(text):
    for key in contractions:
        text = text.replace(key, contractions[key])
    return text

df['description'] = df['description'].apply(expand_contractions)
````

#### 2. Lower Casing

````python
# Lower Casing
df['description'] = df['description'].str.lower()
````

#### 3. Removing Punctuations

````python
# Remove Punctuations
df['description'] = df['description'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
````

#### 4. Removing URLs & Removing words and digits contain digits.

````python
# Remove URLs
df['description'] = df['description'].apply(lambda x: re.sub(r'http\S+', '', x))

# Remove words containing digits
df['description'] = df['description'].apply(lambda x: re.sub(r'\w*\d\w*', '', x))
````

#### 5. Removing Stopwords & Removing White spaces

````python
stop_words = set(stopwords.words('english'))

df['description'] = df['description'].apply(
    lambda x: " ".join([word for word in x.split() if word not in stop_words])
)
````

````python
df['description'] = df['description'].apply(lambda x: re.sub(r'\s+', ' ', x).strip())
````

#### 6. Rephrase Text

````python
# Rephrase Text (basic cleaning applied already)
df['description_cleaned'] = df['description']
````

#### 7. Tokenization

````python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
````

````python
# Tokenization
df['tokens'] = df['description_cleaned'].apply(word_tokenize)
````

#### 8. Text Normalization

````python
lemmatizer = WordNetLemmatizer()

df['normalized_text'] = df['tokens'].apply(
    lambda words: [lemmatizer.lemmatize(word) for word in words]
)
````

##### Which text normalization technique have you used and why?

Lemmatization was used for text normalization.
Unlike stemming, lemmatization converts words into their meaningful base form while preserving context.
It was chosen because it produces linguistically correct root words, improving the quality of text representation for analysis and modeling.

#### 9. Part of speech tagging

````python
nltk.download('averaged_perceptron_tagger_eng')
````

````python
# POS Tagging
df['pos_tags'] = df['normalized_text'].apply(pos_tag)
````

#### 10. Text Vectorization

````python
# Convert list back to string
df['final_text'] = df['normalized_text'].apply(lambda x: " ".join(x))

# TF-IDF Vectorization
vectorizer = TfidfVectorizer(max_features=5000)
X_text = vectorizer.fit_transform(df['final_text'])

X_text.shape
````

##### Which text vectorization technique have you used and why?

TF-IDF (Term Frequency–Inverse Document Frequency) was used for text vectorization.
It assigns importance to words based on their frequency in a document relative to the entire dataset.

TF-IDF was chosen because it reduces the influence of commonly occurring words and highlights more meaningful terms, making it suitable for clustering and machine learning models.

### 5. Data Transformation

#### Do you think that your data needs to be transformed? If yes, which transformation have you used. Explain Why?

Yes, data transformation was required because numerical features such as release_year and content_age have different scales.

Without transformation, features with larger magnitude may dominate the model.

````python
# Transform Your data
import numpy as np

# 1. Prevent NaNs by clipping any accidental negative ages to 0
X_numeric['content_age'] = np.clip(X_numeric['content_age'], 0, None)

# 2. Apply Log Transform to handle skewness
X_numeric['content_age'] = np.log1p(X_numeric['content_age'])

# 3. A bulletproof catch-all to fill any remaining sneaky NaNs
X_numeric.fillna(X_numeric.median(), inplace=True)
````

## ***7. ML Model Implementation***

### ML Model - KMeans Clustering

###1. Explain the ML Model Used

KMeans clustering algorithm was used.

KMeans partitions data into K clusters by minimizing within-cluster variance (WCSS). It assigns each data point to the nearest centroid and updates centroids iteratively until convergence.

It is suitable for:

High-dimensional TF-IDF text features

Unsupervised grouping

Content similarity analysis

###2. Model Evaluation Using Score Chart

Since clustering is unsupervised, traditional metrics such as Accuracy, Precision, Recall, and F1-score are not applicable.

Instead, the following evaluation methods were used:

###🔹 Elbow Method (WCSS)

````python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []
K_range = range(2, 11)

# This loop actually calculates the WCSS for each k!
for k in K_range:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X_reduced)
    wcss.append(kmeans.inertia_)

# Plotting the Elbow Curve
plt.figure(figsize=(10, 5))
plt.plot(K_range, wcss, marker='o', linestyle='--', color='blue')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
````

###🔹 Silhouette Score

````python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

# 1. Apply K-Means with 5 clusters (chosen based on the Elbow Method)
optimal_k = 5
final_kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42)
df['cluster'] = final_kmeans.fit_predict(X_reduced)

# 2. Evaluation Metrics
sil_score = silhouette_score(X_reduced, df['cluster'])
db_score = davies_bouldin_score(X_reduced, df['cluster'])

print(f"Silhouette Score: {sil_score:.4f}")
print(f"Davies-Bouldin Score: {db_score:.4f}")

# 3. Cluster Profiling for Business Insights
print("\n" + "="*50)
print("CLUSTER PROFILING & BUSINESS IMPLICATIONS")
print("="*50)

for i in range(optimal_k):
    print(f"\n--- Cluster {i} ---")
    cluster_data = df[df['cluster'] == i]
    print(f"Total Titles: {len(cluster_data)}")
    print(f"Top 3 Genres:\n{cluster_data['listed_in'].value_counts().head(3).to_string()}")
    print(f"Most Common Rating: {cluster_data['rating'].mode()[0]}")
````

###🔹 Silhouette Score Interpretation
The final model achieved a Silhouette Score of 0.4029 and a Davies-Bouldin Score of 0.8876. In high-dimensional text clustering, a Silhouette Score above 0.4 indicates strong, meaningful separation between the clusters. The Davies-Bouldin score (being under 1.0) further validates that our clusters are dense and distinctly separated.
